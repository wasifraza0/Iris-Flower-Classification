# PDF AND CDF

--> PDF for the iris ML problem: In the iris ML problem, the PDF can be used to visualize the distribution of the features (e.g. sepal length, sepal width, petal length, petal width) of the iris flowers. The PDF can help identify the most likely range of values for each feature, and can be useful in identifying outliers or skewness in the data.

--> CDF for the iris ML problem: The CDF can be used in the iris ML problem to visualize the cumulative distribution of the features, providing information on the proportion of observations that have values less than or equal to a certain value. This can be useful in understanding the distribution of the features and making predictions about future observations.

```python
counts,bin_edges=np.histogram(df['PetalLengthCm'],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
print(pdf)
print(cdf)
```

```python
plt.plot(bin_edges[1:],cdf,label='cdf')
plt.plot(bin_edges[1:],pdf,label='pdf')
plt.legend()
```

```python
counts,bin_edges=np.histogram(df['PetalWidthCm'],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
print(pdf)
print(cdf)
```

```python
plt.plot(bin_edges[1:],cdf,label='cdf')
plt.plot(bin_edges[1:],pdf,label='pdf')
plt.legend()
```

# Box Plot

The box plot consists of a box that represents the interquartile range (IQR) of the data, which is the range between the first quartile (25th percentile) and the third quartile (75th percentile). The box also displays the median of the data as a horizontal line within the box.

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.boxplot(x='Species',y='PetalLengthCm',data=df)
plt.subplot(2,2,2)
sns.boxplot(x='Species',y='PetalWidthCm',data=df)
plt.subplot(2,2,3)
sns.boxplot(x='Species',y='SepalLengthCm',data=df)
plt.subplot(2,2,4)
sns.boxplot(x='Species',y='SepalWidthCm',data=df)
```

# Violin Plot

The violin plot displays the distribution of the data by drawing a symmetrical violin shape around the central part of the distribution. The width of the violin at each point represents the density of the data at that point. The center of the violin plot displays the median and quartiles of the data, similar to a box plot.

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='PetalLengthCm',data=df)
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='PetalWidthCm',data=df)
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='SepalLengthCm',data=df)
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='SepalWidthCm',data=df)
```

## Model Training

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Species'])
Y = df['Species']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30)
```

# ML Model :- Logistic regression

Logistic Regression is a statistical method for analyzing a dataset in which there are one or more independent variables that determine an outcome. It is used for binary classification problems, where the target variable can have only two possible values (e.g. yes/no, true/false, 0/1). In logistic regression, a logistic function is used to model the relationship between the independent variables and the dependent binary variable. The logistic function outputs a probability value, which is transformed into a binary output (i.e. 0 or 1)

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report

# Train the logistic regression model
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

print(classification_report(y_test, y_pred))
```

# DecisionTree Classifier

A Decision Tree Classifier is a type of machine learning algorithm used for classification tasks. It is a simple and interpretable algorithm that works by recursively splitting the data into smaller subsets based on the values of the input features.

The end result of the algorithm is a tree structure, where each node represents a test on a feature, and each leaf node represents a prediction

```python
from sklearn.tree import DecisionTreeClassifier

# Train a decision tree classifier
model_2 = DecisionTreeClassifier()
model_2.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model_2.predict(x_test)


# Print the classification report
print(classification_report(y_test, y_pred))



```

# RandomForest Classifier

Random Forest Classifier is an ensemble machine learning algorithm used for classification tasks. It is built on the decision tree algorithm, but instead of building a single decision tree, it builds multiple decision trees and combines their predictions.

The idea behind this approach is that each decision tree may have its own biases, but by combining the predictions of multiple trees, the overall bias can be reduced, resulting in improved performance and reduced overfitting.

```python
from sklearn.ensemble import RandomForestClassifier
model_3 = RandomForestClassifier()
model_3.fit(x_train, y_train)
# Make predictions on the test set
y_pred = model_3.predict(x_test)

# Print the classification report
print(classification_report(y_test, y_pred))
```

# SupportVector Classifier

Support Vector Classifier (SVC) is a type of machine learning algorithm used for classification tasks. It is a type of linear classifier that tries to find a hyperplane (a line or a plane in high-dimensional space) that separates the data points into different classes. The hyperplane that results in the maximum margin, or the widest separation between the classes, is selected as the final decision boundary.

```python
from sklearn.svm import SVC
model_4= SVC()
model_4.fit(x_train, y_train)
# Make predictions on the test set
y_pred = model_4.predict(x_test)

# Print the classification report
print(classification_report(y_test, y_pred))
```

# KNN Classifier

KNN Classifier (K-Nearest Neighbor Classifier) is a type of machine learning algorithm used for classification tasks. It is a simple and intuitive algorithm that works by finding the K nearest neighbors of a sample in the feature space and assigning the most frequent class among those neighbors as the prediction for that sample.

```python
from sklearn.neighbors import KNeighborsClassifier
model_5 = KNeighborsClassifier(n_neighbors=5)
model_5.fit(x_train, y_train)
# Make predictions on the test set
y_pred = model_4.predict(x_test)
# Print the classification report
print(classification_report(y_test, y_pred))
```

# Hyperparameter tuning

Hyperparameter tuning is the process of selecting the best hyperparameters for a machine learning model. In the case of random forest, the hyperparameters include the number of trees in the forest, the maximum depth of the trees, the minimum number of samples required to split an internal node, and others

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Define the hyperparameters to be tuned
param_grid = {'n_estimators': [100, 200, 300, 400, 500], 'max_depth': [3, 4, 5, 6, 7, 8], 'min_samples_split': [2, 4, 6, 8]}

# Train the random forest model using GridSearchCV
rf = RandomForestClassifier()
rf_cv = GridSearchCV(rf, param_grid, cv=5)
rf_cv.fit(x_train, y_train)

# Print the best hyperparameters
print("Best Hyperparameters:", rf_cv.best_params_)

# Make predictions on the test set
y_pred = rf_cv.predict(x_test)

# Evaluate the performance of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
